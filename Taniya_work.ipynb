```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
from docx import Document
from google.colab import files

# Step 1: Load the Dataset
df = pd.read_csv('The impact of social media on mental health of students in nepal (1).csv')

# Step 2: Create a Word Document for the Analysis
document = Document()
document.add_heading('Analysis of Social Media Impact on Mental Health', level=1)

# Step 3: Dataset Overview
document.add_heading('Dataset Overview', level=2)
document.add_paragraph('Below is the basic structure of the dataset:')
info = df.info(buf=None)
document.add_paragraph(str(info))

# Display first few rows
document.add_heading('Sample Data (First 5 Rows)', level=2)
document.add_paragraph(df.head().to_string())

# Step 4: Handle Missing Data
missing_data = df.isnull().sum()
document.add_heading('Missing Values Summary', level=2)
document.add_paragraph(missing_data.to_string())

# Step 5: Analyze Important Columns
document.add_heading('Key Statistics', level=2)
# Example: Gender Distribution
if 'Gender' in df.columns:
    gender_counts = df['Gender'].value_counts()
    document.add_paragraph('Gender Distribution:')
    document.add_paragraph(gender_counts.to_string())

# Example: Age Statistics
if 'Age' in df.columns:
    age_statistics = df['Age'].describe()
    document.add_paragraph('Age Statistics:')
    document.add_paragraph(age_statistics.to_string())

# Step 6: Create Visualizations
# Gender Pie Chart
if 'Gender' in df.columns:
    gender_counts.plot(kind='pie', autopct='%1.1f%%', figsize=(6, 6), title='Gender Distribution')
    plt.savefig('gender_distribution.png')
    plt.close()
    document.add_picture('gender_distribution.png', width=5)

# Bar Graph: Time Spent on Social Media
if 'TimeOnSocialMedia' in df.columns:
    df['TimeOnSocialMedia'].value_counts().plot(kind='bar', figsize=(8, 5), title='Time Spent on Social Media')
    plt.xlabel('Hours')
    plt.ylabel('Number of Students')
    plt.savefig('time_on_social_media.png')
    plt.close()
    document.add_picture('time_on_social_media.png', width=5)

# Step 7: Save and Export Word Document
document.save('Social_Media_Impact_Analysis.docx')
print("Analysis complete. Downloading the Word document...")

# Step 8: Download the Word Document
files.download('Social_Media_Impact_Analysis.docx')
```

```python
# Install necessary libraries
!pip install python-docx matplotlib pandas

# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
from docx import Document
from docx.shared import Inches
from google.colab import files

# Upload your CSV file
uploaded = files.upload()

# Load the dataset
file_name = list(uploaded.keys())[0]
data = pd.read_csv(file_name)

# Create a Word document for the report
document = Document()
document.add_heading("Social Media Impact Analysis on Mental Health", level=1)

# Dataset Overview
document.add_heading("Dataset Overview", level=2)
document.add_paragraph("Below is the basic structure of the dataset:")
info = data.info(buf=None)  # Get dataset information
document.add_paragraph(str(info))

# Display the first five rows of the dataset
document.add_heading("Sample Data (First 5 Rows)", level=2)
document.add_paragraph(str(data.head()))

# Descriptive statistics
document.add_heading("Descriptive Statistics", level=2)
stats = data.describe(include='all').transpose()
document.add_paragraph(str(stats))

# Generate Gender Distribution Pie Chart
if 'Gender' in data.columns:
    gender_counts = data['Gender'].value_counts()
    plt.figure(figsize=(6, 6))
    gender_counts.plot(kind='pie', autopct='%1.1f%%', colors=['#ff9999', '#66b3ff'], labels=gender_counts.index)
    plt.title("Gender Distribution")
    plt.ylabel("")
    gender_chart_path = 'gender_distribution.png'
    plt.savefig(gender_chart_path)
    plt.close()

    document.add_heading("Gender Distribution", level=2)
    document.add_paragraph("The pie chart below represents the distribution of genders in the dataset.")
    document.add_picture(gender_chart_path, width=Inches(4))

# Generate Age Group Bar Graph
if 'Age' in data.columns:
    plt.figure(figsize=(10, 6))
    data['Age'].plot(kind='hist', bins=10, color='skyblue', edgecolor='black')
    plt.title("Age Distribution")
    plt.xlabel("Age")
    plt.ylabel("Frequency")
    age_chart_path = 'age_distribution.png'
    plt.savefig(age_chart_path)
    plt.close()

    document.add_heading("Age Distribution", level=2)
    document.add_paragraph("The bar graph below shows the distribution of ages in the dataset.")
    document.add_picture(age_chart_path, width=Inches(5))

# Sentiment Analysis (Example for "Sentiment" column)
if 'Sentiment' in data.columns:
    sentiment_counts = data['Sentiment'].value_counts()
    plt.figure(figsize=(8, 6))
    sentiment_counts.plot(kind='bar', color=['#99cc99', '#ff9999', '#66b3ff'], edgecolor='black')
    plt.title("Sentiment Analysis")
    plt.xlabel("Sentiment")
    plt.ylabel("Frequency")
    sentiment_chart_path = 'sentiment_analysis.png'
    plt.savefig(sentiment_chart_path)
    plt.close()

    document.add_heading("Sentiment Analysis", level=2)
    document.add_paragraph("The bar chart below represents the sentiment analysis of social media impact.")
    document.add_picture(sentiment_chart_path, width=Inches(5))

# Save the document
output_file = 'Social_Media_Impact_Analysis_Report.docx'
document.save(output_file)

# Download the report
files.download(output_file)
```

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
from docx import Document
from docx.shared import Inches

# Load dataset
file_name = 'The impact of social media on mental health of students in nepal (1).csv'
data = pd.read_csv(file_name)

# Initialize Word document
document = Document()

# Verify and clean the 'Age' column
if 'Age' in data.columns:
    # Convert 'Age' to numeric, forcing errors to NaN
    data['Age'] = pd.to_numeric(data['Age'], errors='coerce')

    # Drop rows with missing or invalid 'Age' values
    data = data.dropna(subset=['Age'])

    # Generate histogram for 'Age' distribution
    plt.figure(figsize=(10, 6))
    data['Age'].plot(kind='hist', bins=10, color='skyblue', edgecolor='black')
    plt.title("Age Distribution")
    plt.xlabel("Age")
    plt.ylabel("Frequency")
    age_chart_path = 'age_distribution.png'
    plt.savefig(age_chart_path)
    plt.close()

    # Add the Age Distribution chart to the Word document
    document.add_heading("Age Distribution", level=2)
    document.add_paragraph("The bar graph below shows the distribution of ages in the dataset.")
    document.add_picture(age_chart_path, width=Inches(5))
else:
    document.add_heading("Age Distribution", level=2)
    document.add_paragraph("The 'Age' column was not found in the dataset. No chart was generated.")

# Save the Word document
output_path = "Social_Media_Impact_Report.docx"
document.save(output_path)
print(f"Report saved to {output_path}")
```
